```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
text_representations = ['BoW', 'Word2vec', 'FastText', 'BERT', 'CodeBERT']
classifiers = ['DT', 'KNN', 'SVM', 'RF']
categories = [f"{representation} + {classifier}" for representation in text_representations for classifier in classifiers]

# F1-scores for each combination of text representation and classifier
values = [
    # DT
    71.07, 61.20, 69.6, 66.98, 68.96,
    
    # KNN
    73.82, 64.42, 74.45, 70.61, 74.96,
    
    # SVM
    74.01, 63.61, 74.43, 74.51, 74.20,
    
    # RF
    79.15, 68.02, 77.75, 75.14, 78.06,
    
]




with plt.style.context('bmh'):
    # Set the figure size
    plt.figure(figsize=(18, 10))
    
    barWidth = 0.15
    x = np.arange(len(text_representations))
    
    # Create bar chart for each classifier
    for i, classifier in enumerate(classifiers):
        plt.bar(x + i * barWidth, values[i * len(text_representations):(i + 1) * len(text_representations)], width=barWidth, label=classifier)

    # Add category labels to the x-axis with rotation
    plt.xticks(x + barWidth * 2, text_representations, fontsize=18)

    # Add title and labels
    plt.title('Vulnerability Classification', fontweight='bold', fontsize=24)
    plt.xlabel('NLP Method', fontweight='bold', fontsize=20)
    plt.ylabel('F1-score', fontweight='bold', fontsize=20)
    plt.legend(title='Classifier', fontsize=16, title_fontsize=16) #, loc='upper left', bbox_t
    
    plt.yticks(np.arange(0, 101, 20), fontsize=16)
    
    # Show the plot
    plt.tight_layout() 
    plt.show()
```
